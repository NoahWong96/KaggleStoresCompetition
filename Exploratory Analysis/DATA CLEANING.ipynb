```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Data Cleaning

```python
train = pd.read_csv("../Data/train.csv", index_col = 0, parse_dates = ['date'])
test = pd.read_csv("../Data/test.csv", index_col = 0)
train.head(20)
```

```python
train.shape
trainClean = train.dropna()
trainClean["date"] = pd.to_datetime(trainClean["date"])
trainClean["day_of_week"]= trainClean["date"].dt.day_name()
trainClean["weekend"] = trainClean["day_of_week"].isin(["Saturday", "Sunday"]) 
```

Appending Oil prices to trainClean

```python
oil = pd.read_csv("../Data/oil.csv", names = ["date", "price"],parse_dates = ['date'])
oil = oil.drop(index=0)
oil = oil.dropna()
oil["date"] = pd.to_datetime(oil["date"])
trainClean = trainClean.merge(oil, on="date")
```

Adding Column is_holiday to trainClean

```python
holidays = pd.read_csv("../Data/holidays_events.csv")
holidays["date"] = pd.to_datetime(holidays["date"])
holidaysOnly= holidays[holidays["type"] == "Holiday"]
trainClean = trainClean.merge(holidaysOnly, on="date")
print(holidays["locale"].value_counts())
print(trainClean.loc[23654])
print()
```
